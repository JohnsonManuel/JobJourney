```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

# Function to perform the login with retry mechanism
def perform_login(driver, retries=3, delay=5):
    attempt = 0
    while attempt < retries:
        try:
            # Open the website
            driver.get('https://tvsmsampark.darwinbox.in/ms/candidate/careers/a66f688ce907b0')

            # Wait for the username field to be visible
            username = WebDriverWait(driver, 20).until(
                EC.presence_of_element_located((By.XPATH, "//input[@id='username']"))
            )

            # Locate the password and login button as well
            password = driver.find_element(By.XPATH, "//input[@id='password']")
            login_button = driver.find_element(By.XPATH, "//button[@id='login-button']")

            # Enter credentials and submit
            username.send_keys('your_username')
            password.send_keys('your_password')
            login_button.click()

            # If the login is successful, exit the retry loop
            print("Login successful")
            break

        except Exception as e:
            attempt += 1
            print(f"Attempt {attempt} failed: {e}")
            if attempt < retries:
                print(f"Retrying in {delay} seconds...")
                time.sleep(delay)
            else:
                print("All retry attempts failed.")

# Main code
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument("--incognito")  # Use incognito mode to avoid stored cookies/cache
driver = webdriver.Chrome(options=chrome_options)

try:
    # Attempt to perform the login
    perform_login(driver)

finally:
    # Close the browser after completing the process
    driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
import time

# Function to perform the login with retry mechanism
def perform_login(driver, retries=3, delay=5):
    attempt = 0
    while attempt < retries:
        try:
            # Open the website
            driver.get('https://tvsmsampark.darwinbox.in/ms/candidate/careers/a66f688ce907b0')

            # Print the page title for debugging
            print(f"Page title: {driver.title}")

            # Print the page source to ensure elements exist
            print(driver.page_source[:1000])  # Print first 1000 characters of the page source

            # Wait for the username field to be visible and clickable
            username = WebDriverWait(driver, 20).until(
                EC.element_to_be_clickable((By.XPATH, "//input[@id='username']"))
            )
            print("Username field found")

            # Locate the password and login button
            password = WebDriverWait(driver, 20).until(
                EC.element_to_be_clickable((By.XPATH, "//input[@id='password']"))
            )
            print("Password field found")

            login_button = WebDriverWait(driver, 20).until(
                EC.element_to_be_clickable((By.XPATH, "//button[@id='login-button']"))
            )
            print("Login button found")

            # Use ActionChains to ensure clicks are performed
            actions = ActionChains(driver)

            # Move to the username input field and enter username
            actions.move_to_element(username).click().send_keys('your_username').perform()
            print("Entered username")

            # Move to the password input field and enter password
            actions.move_to_element(password).click().send_keys('your_password').perform()
            print("Entered password")

            # Move to the login button and click
            actions.move_to_element(login_button).click().perform()
            print("Clicked login button")

            # If the login is successful, exit the retry loop
            print("Login successful")
            break

        except Exception as e:
            attempt += 1
            print(f"Attempt {attempt} failed: {e}")
            if attempt < retries:
                print(f"Retrying in {delay} seconds...")
                time.sleep(delay)
            else:
                print("All retry attempts failed.")
                print(driver.page_source[:1000])  # Print first 1000 characters of the page source again for final debug

# Main code
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument("--incognito")  # Use incognito mode to avoid stored cookies/cache
driver = webdriver.Chrome(options=chrome_options)

try:
    # Attempt to perform the login
    perform_login(driver)

finally:
    # Close the browser after completing the process
    driver.quit()
```
